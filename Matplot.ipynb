```python
df= pd.read_csv('popdata1.txt', sep='\t')
```

```python
import numpy as np

-np.log(df['P_VALUE'])
```

```python
df['-logp']= - np.log(df['P_VALUE'])
df
```

```python
# Group the 'regions' together
for region, group_df in df.groupby('REGION'):
    print(region)
```

```python
# Put variants in order by max position in chromosome
running_pos = 0

cumulative_pos = []

for region, group_df in df.groupby('REGION'):
    cumulative_pos.append(group_df['CHR_POS'] + running_pos)
    running_pos+= group_df['CHR_POS'].max()

cumulative_pos
```

```python
# Position of variant relative to whole chromosome

df['cumulative_pos'] = pd.concat(cumulative_pos)
df
```

```python
df['SNP number']=df.index
df
```

```python
import seaborn as sns
%matplotlib inline

sns.relplot(
    data = df,
    x= 'cumulative_pos',
    y= '-logp',
    aspect = 2,
    hue = 'REGION'
)
```
