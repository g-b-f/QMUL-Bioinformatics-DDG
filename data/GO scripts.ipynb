```python
import pandas as pd
import requests

# Read the tsv file into a pandas dataframe
df = pd.read_csv("gwas_trimmed.tsv", sep="\t")

# Extract the list of rsids from the dataframe
rsids = df["SNPS"].tolist()


# Initialize a list to store the gene ontology data
go_data = []

# Iterate over the list of rsids
for rsid in rsids:
    # Construct the API request URL
    url = "https://mygene.info/v3/query?q={}&species=human&fields=go".format(rsid)

    # Make the API request
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Extract the gene ontology data from the response
        gene_info = response.json()
        if "hits" in gene_info and len(gene_info["hits"]) > 0:
            go_terms = gene_info["hits"][0].get("go", [])
            for go_term in go_terms:
                go_data.append({"rsid": rsid, "go_term": go_term})

# Convert the go_data list to a pandas dataframe
go_df = pd.DataFrame(go_data)

# Write the go_df dataframe to a tsv file
go_df.to_csv("go_data.tsv", sep="\t", index=False)
```

```python
import pandas as pd
import requests

# Read the tsv file into a pandas dataframe
df = pd.read_csv("gwas_trimmed.tsv", sep="\t")

# Extract the list of rsids from the dataframe
rsids = df["SNPS"].tolist()


# Initialize a list to store the gene ontology data
go_data = []

# Iterate over the list of rsids
for rsid in rsids:
    # First, get the gene symbol for the rsID
    url = "https://mygene.info/v3/query?q={}&species=human&fields=symbol".format(rsid)
    response = requests.get(url)
    
    if response.status_code == 200:
        gene_info = response.json()
        if "hits" in gene_info and len(gene_info["hits"]) > 0:
            gene_symbol = gene_info["hits"][0].get("symbol", None)
            
            # If a gene symbol was found, get the gene ontology information
            if gene_symbol:
                # Construct the URL for the QuickGO API request
                url = "https://www.ebi.ac.uk/QuickGO/services/annotation/search?geneProductId={}&limit=100".format(gene_symbol)
                
                # Make the API request
                response = requests.get(url)
                
                # Check if the request was successful
                if response.status_code == 200:
                    # Extract the gene ontology data from the response
                    gene_ontology = response.json()
                    if "results" in gene_ontology:
                        for result in gene_ontology["results"]:
                            go_data.append({"rsid": rsid, "go_term": result["goName"], "go_id": result["goId"], "go_aspect": result["aspect"], "gene_symbol": gene_symbol})

# Convert the go_data list to a pandas dataframe
go_df = pd.DataFrame(go_data)

# Write the go_df dataframe to a tsv file
go_df.to_csv("go_data.tsv", sep="\t", index=False)
```

```python
import pandas as pd
import requests

# Read the tsv file into a pandas dataframe
df = pd.read_csv("gwas_trimmed.tsv", sep="\t")

# Extract the list of rsids from the dataframe
rsids = df["SNPS"].tolist()

# Initialize a list to store the gene ontology data
go_data = []

# Iterate over the list of rsids
for rsid in rsids:
    # Construct the API request URL for mapping rsids to gene symbols
    url = "https://api.ncbi.nlm.nih.gov/variation/v0/beta/rsvalues?rsids={}".format(rsid)

    # Make the API request
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Extract the gene symbol from the response
        gene_info = response.json()
        if "records" in gene_info and len(gene_info["records"]) > 0:
            gene_symbol = gene_info["records"][0].get("gene_symbol", None)
            if gene_symbol is not None:
                # Construct the API request URL for retrieving gene ontology data
                url = "https://www.ebi.ac.uk/QuickGO/services/annotation/search?geneProductId={}".format(gene_symbol)

                # Make the API request
                response = requests.get(url)

                # Check if the request was successful
                if response.status_code == 200:
                    # Extract the gene ontology data from the response
                    gene_info = response.json()
                    if "results" in gene_info:
                        go_terms = gene_info["results"]
                        for go_term in go_terms:
                            go_data.append({"rsid": rsid, "go_term": go_term})

# Convert the go_data list to a pandas dataframe
go_df = pd.DataFrame(go_data)

# Write the go_df dataframe to a tsv file
go_df.to_csv("go_data.tsv", sep="\t", index=False)
```

```python
import pandas as pd
from Bio import Entrez
from Bio.Entrez import efetch

# Read the tsv file into a Pandas DataFrame
df = pd.read_csv("filtered_go_data2.tsv", sep='\t')

# Use the efetch function from the Bio.Entrez module to retrieve the GO functions
Entrez.email = "apoorvasarah@gmail.com" # Replace with your email address

# Set the batch size for the records to retrieve in a single query
batch_size = 100

# Initialize an empty list to store the GO functions
go_functions = []

# Iterate over the go_id's in batches
for i in range(0, len(df), batch_size):
    # Get the batch of go_id's
    batch = df['go_id'][i:i + batch_size]

    # Use efetch to retrieve the corresponding gene records
    handle = Entrez.efetch(db="gene", id=batch, retmode="xml")
    records = Entrez.read(handle)

    # Extract the GO functions from the records and add them to the list
    go_functions.extend([records[j]['Entrezgene_gene']['Gene-ref'].get('Gene-ref_desc', None) for j in range(len(records))])

# Add the GO functions to the DataFrame as a new column
df['go_function'] = go_functions

# Save the updated DataFrame to a tsv file
df.to_csv("filtered_go_data.tsv", sep='\t', index=False)
```

```python
import requests
import pandas as pd

# Read the tsv file into a Pandas DataFrame
df = pd.read_csv("filtered_go_data.tsv", sep='\t')

# Define the QuickGO API endpoint and headers
quickgo_endpoint = "https://www.ebi.ac.uk/QuickGO/services/annotation/search"
headers = { "Accept": "application/json" }

# Function to get the GO term annotations from the QuickGO API
def get_go_term_annotations(go_id):
    params = {
        "goid": go_id,
        "limit": 1
    }
    response = requests.get(quickgo_endpoint, headers=headers, params=params)
    if response.status_code == 200:
        result = response.json()
        return result["data"][0]["goName"] if len(result["data"]) > 0 else None
    else:
        raise Exception("Failed to retrieve GO term annotations: " + response.text)

# Extract the GO functions from the QuickGO API and store them in a new column in the DataFrame
df['go_function'] = df['go_id'].apply(get_go_term_annotations)
df.to_csv("filtered_go_data.tsv", sep='\t', index=False)
```

```python
import requests
import csv

# Read the input GO IDs from a file
go_ids = []
with open("filtered_go_data.tsv", "r") as f:
    reader = csv.reader(f, delimiter="\t")
    for row in reader:
        go_ids.append(row[0])

# Store the results for each GO ID
results = []
for go_id in go_ids:
    response = requests.get(f"http://api.geneontology.org/api/ontology/go/terms/GO:{go_id}")
    if response.status_code == 200:
        data = response.json()
        results.append([go_id, data["name"], data["definition"]])

# Write the results to a file
with open("output.tsv", "w", newline="") as f:
    writer = csv.writer(f, delimiter="\t")
    writer.writerows(results)
```
